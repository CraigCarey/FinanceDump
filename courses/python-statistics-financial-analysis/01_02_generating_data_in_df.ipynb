```python
import matplotlib as plt
import pandas as pd
pd.__version__

%matplotlib inline
```

```python
fb = pd.read_csv('data/facebook.csv', index_col=0)
```

```python
# Select a single column from the df
fb['Close']

# or
fb.Close
```

```python
# Select multiple columns
fb[['Open', 'Close']]
```

```python
# Create a new column
fb['Price1'] = fb['Close'].shift(-1) # Shift the whole column upwards by 1

display(fb)
```

```python
fb['PriceDiff'] = fb['Price1'] - fb['Close']

display(fb)
```

```python
fb ['Return'] = fb['PriceDiff']/fb['Close']

display(fb)
```

```python
# Use list comprehension when we can't use column-wise transformation
fb['Direction'] = [1 if fb.loc[ei, 'PriceDiff'] > 0 else -1 for ei in fb.index]

display(fb)
```

```python
# Compute rolling average directly

fb['Average3'] = (fb.Close + fb.Close.shift(1) + fb.Close.shift(2))/3

display(fb)
```

```python
# Get the df to compute rolling averages
fb['MA40'] = fb.Close.rolling(40).mean()
fb['MA200'] = fb.Close.rolling(200).mean()

display(fb)
```

```python
# Some momentum traders believe that if MA40 > MA200 then the stock will continue to rise for a while

fb.Close.plot()
fb.MA40.plot()
fb.MA200.plot()
```
