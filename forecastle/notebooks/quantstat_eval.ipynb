```python
%matplotlib inline
import quantstats as qs

# extend pandas functionality with metrics, etc.
qs.extend_pandas()

ticker_code = "VEUR.L"

# fetch the daily returns for a stock
stock = qs.utils.download_returns(ticker_code, period="2y")

# show sharpe ratio
qs.stats.sharpe(stock)
```

```python
# or using extend_pandas() :)
stock.sharpe()
```

```python
print(type(stock))
stock
```

```python
qs.plots.snapshot(stock, title=f'{ticker_code} Performance')
```

```python
qs.reports.html(stock, ticker_code, title=f"{ticker_code} Tearsheet", output=f"tearsheets/{ticker_code}-tearsheet.html")
```
