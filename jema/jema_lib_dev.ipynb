```python
from jema import (
    get_latest_holdings,
    get_symbols_and_exchanges,
    append_symbols_and_exchanges,
    get_local_prices,
    get_fx_rates,
    apply_rates,
)
```

```python
latest_holdings = get_latest_holdings()
display(latest_holdings.head())
```

```python
symbols_and_exchanges = get_symbols_and_exchanges()
display(symbols_and_exchanges.head())
```

```python
latest_holdings_with_symbols = append_symbols_and_exchanges(latest_holdings)
display(latest_holdings_with_symbols.head())
```

```python
# loop until failures is empty
while True:
    failures = get_local_prices(latest_holdings_with_symbols)
    if not failures:
        break

display(failures)
display(latest_holdings_with_symbols.head())
```

```python
fx_rates = get_fx_rates(latest_holdings_with_symbols)
display(fx_rates)
```

```python
apply_rates(latest_holdings_with_symbols, fx_rates)
display(latest_holdings_with_symbols.head())
```
