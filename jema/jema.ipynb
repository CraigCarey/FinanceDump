```python
import pandas as pd
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
# df = pd.read_excel('https://am.jpmorgan.com/content/dam/jpm-am-aem/emea/gb/en/supplemental/full-portfolio-listing/jpm-emerging-europe-middle-east-afria-disclosure.xlsx', sheet_name='June 2024')
jun24 = pd.read_excel('./jpm-emerging-europe-middle-east-afria-disclosure.xlsx', sheet_name='June 2024', skiprows=9, nrows=116)
jun24 = jun24.dropna()
jun24.rename(columns={'Holding': 'Holding Jun24'}, inplace=True)
jun24
```

```python
may24 = pd.read_excel('./jpm-emerging-europe-middle-east-afria-disclosure.xlsx', sheet_name='May 2024', skiprows=9, nrows=116)
may24 = may24.dropna()
may24.rename(columns={'Holding': 'Holding May24'}, inplace=True)
may24
```

```python
df_merged = jun24.merge(may24[['Holding May24', 'Security No.']], on='Security No.', how='left')
df_merged = df_merged[[col for col in df_merged.columns if col != 'Holding Jun24'] + ['Holding Jun24']]
df_merged.fillna(0, inplace=True)
df_merged['Change'] = df_merged['Holding Jun24'] - df_merged['Holding May24']
df_merged
```
