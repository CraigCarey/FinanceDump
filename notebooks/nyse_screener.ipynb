```python
import pickle
import pandas as pd
```

```python
filename = '../data/all_nyse_stock_data_230108.pkl'

with open(filename, 'rb') as file:
    raw_data = pickle.load(file)
```

```python
symbol = 'IMB.L'
symbol_data = raw_data[symbol]
keys = list(symbol_data.keys())
keys
```

```python
headers = list(symbol_data['key_metrics'].index)
headers.insert(0, 'symbol')
# display(headers)

current_data = symbol_data['key_metrics']['2022']
display(current_data)
```

```python
data_dict = current_data.to_dict()
data_dict['symbol'] = symbol
display(data_dict)

df = pd.DataFrame(columns=headers)
display(df)

df2 = pd.concat([df, pd.DataFrame([data_dict])], ignore_index=True)
display(df2)
```
