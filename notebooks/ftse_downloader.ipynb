```python
import pandas as pd
```

```python
ticker_files = [
'ftse-100',
'ftse-250',
'ftse-350',
'ftse-aim-100',
'ftse-all-share',
'ftse-small-cap',
'ftse-techmark-100']

ticker_symbols = []

for f in ticker_files:
    print(f"reading ../scraping/{f}.csv")
    df = pd.read_csv(f"../scraping/{f}.csv")
    df_tickers = df.index.values.tolist()
    ticker_symbols.extend(df_tickers)

ticker_symbols = sorted(set(ticker_symbols))
print("Number after cleaning:", len(ticker_symbols))
```

```python
import yfinance as yf
import pickle

file = open('all_ftse_221210.pkl', 'wb')

tickers_dict = {}

cnt = 0

for ticker_symbol in ticker_symbols:
    ticker_symbol += ".L"
    print(f"pickling {ticker_symbol}")
    ticker = yf.Ticker(ticker_symbol)
    if ('symbol' not in ticker.info):
        print(f"Failed to get data for {ticker_symbol}")
        continue
    tickers_dict[ticker_symbol] = ticker

pickle.dump(tickers_dict, file)
file.close()
```

```python
for ticker_symbol in ticker_symbols:
    if 'BT.A' in ticker_symbol:
        ticker_symbol = 'BT-A.L'
    elif '.' in ticker_symbol:
        ticker_symbol += "L"
    else:
        ticker_symbol += ".L"
    
    if ticker_symbol not in tickers_dict:
        print(f"need {ticker_symbol}")
        ticker = yf.Ticker(ticker_symbol)
        if ('symbol' not in ticker.info):
            print(f"Failed to get data for {ticker_symbol}")
            continue
        tickers_dict[ticker_symbol] = ticker
```
