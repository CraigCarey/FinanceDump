```python
import yfinance as yf
from datetime import datetime 
```

```python
symbol = "IMB.L"
ticker = yf.Ticker(symbol)
```

```python
ticker.
```

```python
sector = ticker.info['sector']
industry = ticker.info['industry']
div_yield = ticker.info['dividendYield']
div_payout_ratio = ticker.info['payoutRatio']
five_yr_div_yld_avg = ticker.info['fiveYearAvgDividendYield']
ex_div_date = datetime.utcfromtimestamp(ticker.info['exDividendDate']).strftime('%y-%m-%d')
pe_ttm = ticker.info['trailingPE']
pe_ntm = ticker.info['forwardPE']
earnings_growth_rate = 0.098 #ticker.info['earningsGrowth']
pegy_ratio_ttm = pe_ttm / (earnings_growth_rate*100 + div_yield*100)
pegy_ratio_ntm = pe_ntm / (earnings_growth_rate*100 + div_yield*100)

print(f'{symbol}')
print(f'\tsector: {sector}')
print(f'\tindustry: {industry}')
print()
print(f'\tdiv_yield: {div_yield*100:0.1f}%')
print(f'\tfive_yr_div_yld_avg: {five_yr_div_yld_avg:0.0f}%')
print(f'\tpayout ratio: {div_payout_ratio*100:0.0f}%')
print(f'\tex-dividend date: {ex_div_date}')
print(f'\tpe_ttm: {pe_ttm}')
print(f'\tpe_ntm: {pe_ntm}')
print(f'\tpegy_ratio_ttm: {pegy_ratio_ttm:0.2f}')
print(f'\tpegy_ratio_ntm: {pegy_ratio_ntm:0.2f}')
```
