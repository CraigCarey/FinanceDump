```python
available_companies_file = f'../data/all_available_companies_230429.pkl'
with open(available_companies_file, 'rb') as file:
    available_companies_raw = pickle.load(file)

display(available_companies_raw)

types = sorted(available_companies_raw["type"].unique())
display(types)
```
