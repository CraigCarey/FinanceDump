```python
import pandas as pd
import pickle
```

```python
symbol = 'POLY.L'
```

```python
# get symbol data from exchange data file and dump into its own pickle
lse_datafile = '../data/all_lse_stock_data_230108.pkl'

with open(lse_datafile, 'rb') as file:
    lse_data = pickle.load(file)

print(len(lse_data))

# symbol_data = lse_data[symbol]
# display(symbol_data.keys())

# with open(f'../data/fmp_{symbol}.pkl', 'wb') as file:
#     pickle.dump(symbol_data, file)
```

```python
# load the symbol data from pickle
symbol_datafile = f'../data/fmp_{symbol}.pkl'

with open(symbol_datafile, 'rb') as file:
    symbol_data = pickle.load(file)
```

```python
display(symbol_data.keys())
# display(symbol_data['key_metrics'])
# display(symbol_data['key_metrics_quarterly'])
# display(symbol_data['key_metrics_ttm'])

key_metrics_type = 'key_metrics'
```

```python
if key_metrics_type == 'key_metrics_ttm':
    year = 'ttm'
else:
    year = symbol_data[key_metrics_type].columns[0]
display(year)
```

```python
# Market Cap - used for filtering
if key_metrics_type == 'key_metrics_ttm':
    market_cap = symbol_data[key_metrics_type]['marketCapTTM']
else:
    market_cap = symbol_data[key_metrics_type].iloc[:, 0]['marketCap']
display(market_cap)
```

```python
# ROIC: return on invested capital - a measure of quality
#     ROIC = NOPAT / Invested Capital
#     NOPAT = Net Operating Profit After Tax
if key_metrics_type == 'key_metrics_ttm':
    roic = symbol_data[key_metrics_type]['roicTTM']
else:
    roic = symbol_data[key_metrics_type].iloc[:, 0]['roic']
display(roic)
```

```python
# Earnings Yield - A measure of cheapness
#     EBIT / EV (Earnings before Interest and Taxes/Enterprise Value)
if key_metrics_type == 'key_metrics_ttm':
    earnings_yield = symbol_data[key_metrics_type]['earningsYieldTTM']
else:
    earnings_yield = symbol_data[key_metrics_type]['2021']['earningsYield']
display(earnings_yield)
```

```python
mf_dict = {
    'symbol': symbol,
    'year': year,
    'market_cap': market_cap,
    'roic': roic,
    'earnings_yield': earnings_yield
}
```

```python
mf_df = pd.DataFrame(mf_dict, index=[0])
display(mf_df)
```

```python
# mf_df.loc[-1] = [symbol, market_cap, roic, earnings_yield]

mf_df2 = pd.DataFrame(mf_dict, index=[0])
mf_df = pd.concat([mf_df2, mf_df], ignore_index=True)
display(mf_df)
```

```python
mf_df = pd.DataFrame(columns=['symbol','market_cap', 'roic', 'earnings_yield'])

for key in lse_data.keys():
    print(key)
    
    if not key_metrics_type in lse_data[key]:
        print(f'{key_metrics_type} not found')
        continue

    key_metrics = lse_data[key][key_metrics_type]

    if not isinstance(key_metrics, pd.DataFrame):
        print(f'{key_metrics_type} is None')
        continue

    if key_metrics.empty:
        print(f'{key_metrics_type} is Empty')
        continue

    if key_metrics_type == 'key_metrics_ttm':
        # year = 'ttm'
        market_cap = key_metrics['marketCapTTM']
        roic =  key_metrics['roicTTM']
        earnings_yield =  key_metrics['earningsYieldTTM']
    else:
        year = key_metrics.columns[0]
        market_cap = key_metrics[year]['marketCap']
        roic =  key_metrics[year]['roic']
        earnings_yield =  key_metrics[year]['earningsYield']

    if not market_cap or not roic or not earnings_yield:
        print(f'empty val found')
        continue

    symbol_mf_dict = {
        'symbol': key,
        'year': year,
        'market_cap': market_cap,
        'roic': roic,
        'earnings_yield': earnings_yield
    }


    mf_df2 = pd.DataFrame(symbol_mf_dict, index=[0])
    mf_df = pd.concat([mf_df2, mf_df], ignore_index=True)

display(mf_df)
```

```python
mf_df["roic_rank"] = mf_df["roic"].rank(method="dense", ascending=False)
mf_df["ey_rank"] = mf_df["earnings_yield"].rank(method="dense", ascending=False)
mf_df["mf_rank"] = mf_df["roic_rank"] + mf_df['ey_rank']
display(mf_df)

display(mf_df.loc[mf_df['roic_rank'].idxmin()])
display(mf_df.loc[mf_df['ey_rank'].idxmin()])
display(mf_df.loc[mf_df['mf_rank'].idxmin()])
```
